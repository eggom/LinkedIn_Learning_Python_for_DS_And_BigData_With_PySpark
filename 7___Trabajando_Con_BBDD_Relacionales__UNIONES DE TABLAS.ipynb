```python
import pandas as pd
```

```python
con_df = pd.read_excel("./data_src/joins_dummies.xlsx", sheet_name="Tabla1")
con_df
```

```python
prod_df = pd.read_excel("./data_src/joins_dummies.xlsx", sheet_name="Tabla2")
prod_df
```

## Uniendo Tablas con pd.merge

#### OUTER JOIN

```python
pd.merge(con_df,prod_df,on="Producto",how="outer")
```

#### INNER JOIN

```python
pd.merge(con_df,prod_df,on="Producto",how="inner")
```

#### RIGHT JOIN

Júntame la tabla de la izquierda con los de la derecha SÓLO SI sus valores aparecen en la tabla de la derecha

```python
pd.merge(con_df,prod_df,on="Producto",how="right")
```

#### LEFT JOIN

Júntame la tabla de la izquierda con los de la derecha SÓLO SI sus valores aparecen en la tabla de la izquierda

```python
pd.merge(con_df,prod_df,on="Producto",how="left")
```

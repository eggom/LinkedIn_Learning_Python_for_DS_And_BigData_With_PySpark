### Empleando 'groupby' con pandas

Groupby nos ayuda a realizar Tablas de Frecuencias y Tablas con funciones de agregación...

````python
# La sintaxis es sencilla:

# DataFrame.groupby(by=Variable_Dentro_DataFrame_Por_La_Que_Quiero_Agrupar)["Variable_Dentro_DataFrame_Que_Quiero_Observar"].función_de_agregacion()

# Si quiero ver el máximo valor de minutos de retraso de salida de los vuelos según el día de la semana:
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_La/s_Que_Quiero_Agrupar: DayOfWeek
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay
# función_de_agregación(): max()

df.groupby(by="DayOfWeek")["ArrDelay"].max()
````

````python
# La sintaxis es sencilla:

# DataFrame.groupby(by=Variable_Dentro_DataFrame_Por_La_Que_Quiero_Agrupar)["Variable_Dentro_DataFrame_Que_Quiero_Observar"].función_de_agregacion()

# Si quiero ver el número medio de minutos de retraso de salida de los vuelos según el destino:
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_La/s_Que_Quiero_Agrupar: Dest
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay
# función_de_agregación(): mean()

df.groupby(by="Dest")["ArrDelay"].mean()
````

Las funciones de agregación más comunes (para variables numericas) son:

* .max()
* .min()
* .count()
* .sum()
* .mean()
* .stdev()

Y una MUY MUY interesante es .describe() para mostrar un resumen estadistico:

````python
# Si quiero ver el resumen estadístico de los minutos de retraso de salida de los vuelos según el día de la semana:
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_Las_Que_Quiero_Agrupar: DayOfWeek
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay
# función_de_agregación(): describe()

df.groupby(by="DayOfWeek")["ArrDelay"].describe()
````

````python
# Si quiero ver los valores medios de de los minutos de retraso (tanto en la salida como en la llegada) de los vuelos
# según el día de la semana:
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_Las_Que_Quiero_Agrupar: DayOfWeek
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay, DepDelay
# función_de_agregación(): mean()
#
#
# Importante siempre comparar cosas con sentido!!!

df.groupby(by="DayOfWeek")[["ArrDelay","DepDelay"]].mean()
````

#### Interesante en este punto:

El empleo de [groupby.agg](https://pandas.pydata.org/pandas-docs/version/0.23.1/generated/pandas.core.groupby.DataFrameGroupBy.agg.html)

### Operaciones con los objetos 'groupby':

````python
# Si quiero ver el rango de minutos de retraso de salida de los vuelos según el día de la semana
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_Las_Que_Quiero_Agrupar: DayOfWeek
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay
# función_de_agregación(): max() - min()

df.groupby(by="DayOfWeek")["ArrDelay"].max() - df.groupby(by="DayOfWeek")["ArrDelay"].min()
````

````python
# Empleando el dataframe que contiene vuelos que salen de SFO y HOU quiere ver las medias de retraso de salida de los
# vuelos según día de la semana y aeropuerto de salida.
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_Las_Que_Quiero_Agrupar: DayOfWeek, Origin
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay
# función_de_agregación(): mean()
#
# MUY IMPORTANTE elegir bien el orden de variables dentro del by= pues el análisis será muy diferente :)

df_SFOHOU.groupby(by=["DayOfWeek","Origin"])["ArrDelay"].mean()
````

````python
# Se pueden guardar objetos groupby a los que luego es más fácil aplicarles funciones de agregación.
#
# Si sé que realizaré muchos análisis del retraso de vuelos en su llegada por día de la semana puedo definir el objeto
# groupby sin función de agregación
#
# DataFrame: df
# Variable/s_Dentro_DataFrame_Por_Las_Que_Quiero_Agrupar: DayOfWeek, Origin
# Variable/s_Dentro_DataFrame_Que_Quiero_Observar: ArrDelay

mygroupby = df.groupby(by="DayOfWeek")["ArrDelay"]
````

````python
# Luego puedo usar ese objeto 'mygroupby' para pasarle diferentes funciones de agregación:

mygroupby.mean()
````

````python
mygroupby.max()
````

## Gestionar información duplicada o información faltante en los dataframes

### Gestión de datos duplicados.

Salvo casos muy concretos, el procedimiento habitual de actuación es el de eliminar registros duplicados.

En este ejemplo forzaré el tener datos duplicados

````python
# Cargo nuevamente df esta vez con 250.000 filas (que sabemos que no tienen datos repetidos)
df = df = pd.read_csv("./data_src/DelayedFlights.csv", nrows=250000)
````

````python
# Genero un df_duplicate que sea una fracción del 20% de registros de df aleatoriamente escogidos

df_duplicate = df.sample(frac=0.2)
````

````python
# Como df tenía 250.000 observaciones, df_duplicate deberá tener un 20% de estas = 50.000 observaciones. Comprobamos:
df_duplicate.shape
````

````python
# Añadimos este df_duplicate por debajo de df

df = df.append(df_duplicate)
df.shape
````

Ya tenemos nuestro dataframe original (250K obs) + 50K duplicados. Comenzamos:

````python
# Preguntamos a Pandas cuántas obs (filas) hay duplicadas en nuestro DataFrame

df.duplicated().sum()
````

````python
# El dato es coherente con nuestras 50K filas forzadas.

# Para eliminar duplicados es tan sencillo como usar DataFrame.drop_duplicates()

# Usamos inplace si queremos que la acción sea efectiva o no.

df.drop_duplicates(inplace=True)
df.shape
````

#### El poder de subset en .duplicated(subset=···, ···) o .drop_duplicates(subset=···, ···)

```subset=``` permite definir un conjunto, en este caso de columnas, donde buscar si hay duplicados bien para informar (.duplicated) o bien para borrar (.drop_duplicates)

````python
# Sabemos que nuestro df ahora NO TIENE observaciones en las que TODAS sus variables se encuentran duplicadas:

df.duplicated().sum()
````

````python
# Pero, ¿y si queremos saber cuantas observaciones tienen el 'FlightNum' repetido?
df.duplicated(subset="FlightNum").sum()
````

````python
# Y aquellos que tienen FlightNum+Origin+Dest repetido?
df.duplicated(subset=["FlightNum","Origin","Dest"]).sum()
````

````python
# Si ahora decidimos quedarnos con los no repetidos, nos queda un DataFrame de tan solo 31305 observaciones
df.drop_duplicates(subset=["FlightNum","Origin","Dest"])
````

### Gestión de Datos Faltantes (una aproximación muy básica)

Aquí sólo hablaremos de eliminación de observaciones (aka registros, aka filas).

No se hablará del tema de imputación de valores.

````python
# Cargo nuevamente df esta vez con 250.000 filas (que sabemos que no tienen datos repetidos)
df = df = pd.read_csv("./data_src/DelayedFlights.csv", nrows=250000)
````

````python
# Analizamos cuántos datos hay faltantes por columna:

# Manera 1:
# df.info()

# Manera 2:
df.isna().sum()
````

#### Uso de dropna() para eliminar datos faltantes

````python
# Si queremos eliminar todas las observaciones con al menos un NaN entre sus atributos:

df.dropna()
````

````python
# Podemos también decirle a dropna que elimine observaciones con un cierto threshold.
#
# Entender threshold en dropna es un poco complicado. Lo intentaremos explicar.
#
# Para manejar el threshold es importante conocer primero el número total de columnas de nuestro dataframe.
# En nuestro caso: 30

# Ahora comenzamos a jugar con threshold en .dropna(···)

# df.dropna(thresh= total_columnas_dataframe - num_NaN_minimos_del_registro_a_eliminar + 1)

# Si queremos eliminar los registros que contengan 1 NaN o más:
df.dropna(thresh=30-1+1)

# Si queremos eliminar los registros que contengan 4 NaN o más:
df.dropna(thresh=30-4+1)
````

````python
# Usando subset en dropna(···)

# Si queremos eliminar todas las filas que contienen NaN en la columna CarrierDelay entonces:

df.dropna(subset=["CarrierDelay"])
````

# Paralelizando procesos en Python

La librería joblib y sus clases Parallel y delayed

```python
import pandas as pd
import numpy as np

from joblib import Parallel, delayed
```

```python
df = pd.read_csv("./data_src/DelayedFlights.csv", nrows=150000)
```

```python
# Nos fijamos en las 5 últimas columnas del DF:

df.columns.tolist()[-5:]
```

```python
# Subseteamos un df con ese conjunto de columnas:

df_sub = df[df.columns.tolist()[-5:]]
df_sub
```

```python
# Definimos una función a implementar

def retraso_maximo(fila):
    if not np.isnan(fila).any():
        names = df.columns.tolist()[-5:]
        return names[fila.index(max(fila))]
    else:
        return "None"
```

```python
results = []

for fila in df_sub.values.tolist():
    results.append(retraso_maximo(fila))
    
results
```

## Paralelizamos el proceso:

```python
result = Parallel(n_jobs = 3, backend = "multiprocessing")(
            map(delayed(retraso_maximo), df_sub.values.tolist()))
```
